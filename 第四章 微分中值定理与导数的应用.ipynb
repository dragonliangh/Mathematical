                            4.2 洛必达法则

    洛必达法则的基本形式
设函数 f(x)f(x) 和 g(x)g(x) 在点 a 的某个去心邻域内可导，且满足以下条件之一：
limx→af(x)=0 且 limx→ag(x)=0（即 00|00 型不定型）。
limx→af(x)=∞ 且 lim⁡x→ag(x)=∞（即 ∞∞|∞∞ 型不定型）。
如果 limx→af′(x)/g′(x) 存在或为 ∞|∞，则：
limx→af(x)/g(x)=limx→af′(x)/g′(x).

```python
import sympy as sy
x = sy.symbols('x')
y = x**x
lim_ex_1 = sy.limit(y,x,0,dir = '+-')
print(lim_ex_1)
```

                        4.3 泰勒公式

【例4.3】写出函数f(x)=e^x的带有佩亚诺余项的麦克劳林公式

```python
import sympy as sy

x = sy.symbols('x')
y = sy.E**x
expl = sy.series(y,x,0)     # a=0时，泰勒公式称为麦克劳林公式
print(expl)
```

泰勒公式的余项：

1、拉格朗日余项：Rn(x) = f(ξ)^(n+1)/(n+1)!*(x-a)^(n+1)。其中ξ是a和x之间的某个值。

2、佩亚诺余项：Rn(x) = o((x−a)^n)，表示当x->a时，余项比(x-a)^n更快趋近于0。

3、麦克劳林公式：当a = 0时，泰勒公式称为麦克劳林公式

f(x) = f(0)+f'(0)*x+f"(0)x/2!*x^2...+f'n(0)/n!*x^n+Rn(x)

4、常见函数的泰勒展开(麦克劳林展开)

(1)指数函数e^x = ∑(x^k)/k!

(2)正弦函数sin(x) = ∑(-1)^k * x^(2k+1)/(2k+1)!

(3)余弦函数cos(x) = ∑(-1)^k * x^2k/2k!

(4)自然对数ln(1+x) = ∑(-1)^(k+1) * x^k/k，|x|<1

(5)几何级数1/(1-x) = ∑x^k，|x|<1。

泰勒公式的应用主要有几个方面的：1、函数近似值，2、极限计算，3、数值分析，4、理论研究比如证明定理和分析函数性质

                        4.4 函数的单调性与曲线的凹凸性

函数的单调性和凹凸性的判定离不开函数的一阶导数和二阶导数，同时结合函数的图像可起到事半功倍的效果

例【4.10】判定函数y = x - sin(x)在[-π，π]上的单调性。

```python
import sympy as sy
x = sy.symbols('x')
f = x - sy.sin(x)
diff_y = sy.diff(f,x,1)
print(diff_y)
```

绘图验证：

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 400)
y = x - np.sin(x)
dy = 1 - np.cos(x)

plt.figure(figsize=(10, 5))

# 绘制函数图像
plt.subplot(1, 2, 1)
plt.plot(x, y, label='$y = x - sin(x)$')
plt.scatter([-np.pi, 0, np.pi], [-np.pi, 0, np.pi], color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function')
plt.grid(True)
plt.legend()

# 绘制导数图像
plt.subplot(1, 2, 2)
plt.plot(x, dy, label="$y' = 1 - cos(x)$")
plt.scatter([-np.pi, 0, np.pi], [2, 0, 2], color='red')
plt.xlabel('x')
plt.ylabel("y'")
plt.title('Derivative')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

由于函数y在[-π，π]上连续，在(-π，π)内导函数1-cos(x)≥0，且等号仅在x=0处成立，所以函数在[-π，π]上单调递增。

【例4.12】求y=2*x^3+3*x^2-12*x+14的拐点。

```python
import sympy as sy
x = sy.symbols('x')
y = 2*x**3+3*x**2-12*x+14
diff_1 = sy.diff(y,x,1)
diff_2 = sy.diff(diff_1,x,1)
inflextion_x = sy.solve(diff_2,x)
print(inflextion_x)
inflextion_y = y.subs(x,inflextion_x[0])
print(inflextion_y)
```

                        4.5 函数的极值与最大最小值

引入scipy.optimize模块的fmin()函数，找到函数的局部极小值点。

【例4.13】讨论函数y = x + 3sin(x)的极值。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin  # 导入 fmin 函数

# 定义目标函数
def func(x):
    return x + 3 * np.sin(x)

# 生成数据点（用于绘图）
x = np.linspace(0, 50, 401)
y = func(x)

# 绘制函数曲线
plt.figure(figsize=(10, 5))
plt.plot(x, y, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# 使用 fmin 寻找局部极小值（需指定初始猜测点）
initial_guess = 2  # 初始猜测值（例如，根据图形选择一个附近的点）
result = fmin(func, initial_guess, disp=False)  # disp=False 关闭冗余输出

# 提取结果并标注在图上
min_x = result[0]
min_y = func(min_x)
plt.scatter(min_x, min_y, color='blue', zorder=5, label=f'Local Minima at x≈{min_x:.4f}')
plt.legend()

plt.show()
```

【例4.14】求函数$f(x) = (x^2 - 1)^3 + 1$的极值。

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**2 - 1)**3 + 1

x = np.linspace(-1.5, 1.5, 400)
y = f(x)

critical_points = [-1, 0, 1]
colors = ['red', 'blue', 'red']  # 颜色区分极值点和其他临界点

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$f(x) = (x^2 - 1)^3 + 1$')

# 标出临界点并注释极值
for point, color in zip(critical_points, colors):
    plt.scatter(point, f(point), color=color, zorder=3)

plt.annotate('Local Minimum (0, 0)', xy=(0, 0), xytext=(0.5, -0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             horizontalalignment='center')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Extrema of $f(x) = (x^2 - 1)^3 + 1$')
plt.grid(True)
plt.legend()
plt.show()
```

【例4.15】求函数 $f(x) = (3x^2+4x+4)/(x^2+x+1)$的极值

```python
import sympy as sy

x = sy.symbols('x')
f = (3*x**2 + 4*x + 4)/(x**2 + x + 1)

# 求一阶导数
f_prime = sy.diff(f, x)
critical_points = sy.solve(f_prime, x)  # 解得 x = 0, x = -2

# 计算极值
extremum_values = [(point, f.subs(x, point)) for point in critical_points]
extremum_types = []

# 判断极值类型（符号变化法）
for point in critical_points:
    left = f_prime.subs(x, point - 1e-5)
    right = f_prime.subs(x, point + 1e-5)
    if left < 0 and right > 0:
        extremum_types.append("极小值")
    elif left > 0 and right < 0:
        extremum_types.append("极大值")
    else:
        extremum_types.append("非极值")

# 输出结果
for (point, value), etype in zip(extremum_values, extremum_types):
    print(f"{etype}点：({point}, {value.evalf()})")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数
def f(x):
    return (3 * x**2 + 4 * x + 4) / (x**2 + x + 1)

# 生成数据点
x = np.linspace(-5, 3, 500)
y = f(x)

# 极值点坐标
max_point = (0, 4)
min_point = (-2, 8/3)

plt.figure(figsize=(10, 6))

# 绘制函数曲线
plt.plot(x, y, label=r'$f(x) = \frac{3x^2 + 4x + 4}{x^2 + x + 1}$', color='blue')

# 标记极值点
plt.scatter(*max_point, color='red', s=100, zorder=5, label='Local Maximum (0, 4)')
plt.scatter(*min_point, color='green', s=100, zorder=5, label='Local Minimum (-2, 8/3)')

# 水平渐近线 y=3
plt.axhline(y=3, color='gray', linestyle='--', alpha=0.5, label='Horizontal Asymptote y=3')

# 标注文本
plt.text(0.2, 4.1, 'Maximum (0, 4)', color='red', ha='left')
plt.text(-2.3, 2.4, 'Minimum (-2, 8/3)', color='green', ha='left')

# 设置图形属性
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function and Extrema')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.ylim(0, 5)  # 限制y轴范围以突出极值
plt.show()
```
