                            4.2 洛必达法则

    洛必达法则的基本形式
设函数 f(x)f(x) 和 g(x)g(x) 在点 a 的某个去心邻域内可导，且满足以下条件之一：
limx→af(x)=0 且 limx→ag(x)=0（即 00|00 型不定型）。
limx→af(x)=∞ 且 lim⁡x→ag(x)=∞（即 ∞∞|∞∞ 型不定型）。
如果 limx→af′(x)/g′(x) 存在或为 ∞|∞，则：
limx→af(x)/g(x)=limx→af′(x)/g′(x).

```python
import sympy as sy
x = sy.symbols('x')
y = x**x
lim_ex_1 = sy.limit(y,x,0,dir = '+-')
print(lim_ex_1)
```

                        4.3 泰勒公式

【例4.3】写出函数f(x)=e^x的带有佩亚诺余项的麦克劳林公式

```python
import sympy as sy

x = sy.symbols('x')
y = sy.E**x
expl = sy.series(y,x,0)     # a=0时，泰勒公式称为麦克劳林公式
print(expl)
```

泰勒公式的余项：

1、拉格朗日余项：Rn(x) = f(ξ)^(n+1)/(n+1)!*(x-a)^(n+1)。其中ξ是a和x之间的某个值。

2、佩亚诺余项：Rn(x) = o((x−a)^n)，表示当x->a时，余项比(x-a)^n更快趋近于0。

3、麦克劳林公式：当a = 0时，泰勒公式称为麦克劳林公式

f(x) = f(0)+f'(0)*x+f"(0)x/2!*x^2...+f'n(0)/n!*x^n+Rn(x)

4、常见函数的泰勒展开(麦克劳林展开)

(1)指数函数e^x = ∑(x^k)/k!

(2)正弦函数sin(x) = ∑(-1)^k * x^(2k+1)/(2k+1)!

(3)余弦函数cos(x) = ∑(-1)^k * x^2k/2k!

(4)自然对数ln(1+x) = ∑(-1)^(k+1) * x^k/k，|x|<1

(5)几何级数1/(1-x) = ∑x^k，|x|<1。

泰勒公式的应用主要有几个方面的：1、函数近似值，2、极限计算，3、数值分析，4、理论研究比如证明定理和分析函数性质

                        4.4 函数的单调性与曲线的凹凸性

函数的单调性和凹凸性的判定离不开函数的一阶导数和二阶导数，同时结合函数的图像可起到事半功倍的效果

例【4.10】判定函数y = x - sin(x)在[-π，π]上的单调性。

```python
import sympy as sy
x = sy.symbols('x')
f = x - sy.sin(x)
diff_y = sy.diff(f,x,1)
print(diff_y)
```

绘图验证：

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 400)
y = x - np.sin(x)
dy = 1 - np.cos(x)

plt.figure(figsize=(10, 5))

# 绘制函数图像
plt.subplot(1, 2, 1)
plt.plot(x, y, label='$y = x - sin(x)$')
plt.scatter([-np.pi, 0, np.pi], [-np.pi, 0, np.pi], color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function')
plt.grid(True)
plt.legend()

# 绘制导数图像
plt.subplot(1, 2, 2)
plt.plot(x, dy, label="$y' = 1 - cos(x)$")
plt.scatter([-np.pi, 0, np.pi], [2, 0, 2], color='red')
plt.xlabel('x')
plt.ylabel("y'")
plt.title('Derivative')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

由于函数y在[-π，π]上连续，在(-π，π)内导函数1-cos(x)≥0，且等号仅在x=0处成立，所以函数在[-π，π]上单调递增。

【例4.11】讨论函数y = (x^2)^(1/3)的单调性。

```python
import sympy as sy
x = sy.symbols('x')
f = (x**2)**(1/3)
diff_y = sy.diff(f,x,1)
print(diff_y)
```

绘图观察

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2,2,201)
plt.plot(x,(x**2)**(1/3),label = r'$y = x**(2/3)$',color = 'red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curve of Ex_3')
plt.ylim(0,2)        #设置y轴坐标范围为0~2.注意书上的代码plt.axes().set_ylim(0,2)是错的。
plt.legend()
plt.grid(True)
plt.show()
```

从输出结果可以看到，函数在(-∞，0)单调递减，在(0，+∞)单调递增

【例4.12】求y=2*x^3+3*x^2-12*x+14的拐点。

```python
import sympy as sy
x = sy.symbols('x')
y = 2*x**3+3*x**2-12*x+14
diff_1 = sy.diff(y,x,1)
diff_2 = sy.diff(diff_1,x,1)
inflextion_x = sy.solve(diff_2,x)
print(inflextion_x)
inflextion_y = y.subs(x,inflextion_x[0])
print(inflextion_y)
```

绘制图像验证：

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,2,401)
y = 2*x**3+3*x**2-12*x+14
y_second_derivative = 12*x + 6

plt.figure(figsize=(10, 5))

# 绘制原函数
plt.subplot(1, 2, 1)
plt.plot(x,y,label = r'y = 2x^3+3X^2-12x+14',color = 'red')
plt.scatter(-0.5,20.5,color = 'blue',zorder = 5 ,label = 'nflection point')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# 绘制二阶导数
plt.subplot(1, 2, 2)
plt.plot(x,y_second_derivative,label=r'$y = 12x+6$',color = 'red')
plt.axhline(0,color='black',linestyle='--')
plt.scatter(-0.5,0,color = 'blue',zorder = 5 ,label = 'second derivative')
plt.xlabel('x')
plt.ylabel(r"$y''$")
plt.legend()
plt.grid(True)
plt.show()
```
