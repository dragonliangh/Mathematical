从数据看，函数逐渐向1趋近，那么看看更大的数据绘图情况。
极限的证明：当n>N时，a(n)全部落在y=1-e与y=1+e之间，即|a-1|<e,故数列极限为1。


```python
import numpy as np
import matplotlib.pyplot as plt

def f(n):
    return (n+np.power(-1,n-1))/n

x = []
for i in range(90,160):
    x.append(i)
y = []
for i in range(len(x)):
    y.append(f(x[i]))
epsilon = 0.01
fig,ax = plt.subplots()
# 绘制水平参考线
ax.hlines(1-epsilon,90,160,'r',label='y=0.99',linewidth=2,linestyle='--')
ax.hlines(1+epsilon,90,160,'r',label='y=1.01',linewidth=2,linestyle='--')
ax.hlines(1,90,160,'k',label='y=1.00',linewidth=2,linestyle='--')
# 绘制散点图
ax.scatter(x,y,c='b',s=10,label='f(n)')
# 确定N值，当n>N时，|y-1|<epsilon。
for N in range(len(y)):
    if np.abs(y[N]-1)<epsilon:
        break
N-=1
N+=90
# 设置垂直参考线
ax.vlines(N,0.989,1.011,linestyle='--',label='n=N',color='r')
ax.set_title('Defination limit of a series of numbers by Epsilon')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

【例2.6】求x->1,(x^2)/(x-1)的极限。

```python
import sympy as sy

x=sy.symbols('x')
def f(x):
    return (x**2-1)/(x-1)

result = sy.limit(f(x),x,1,dir='+-')
```

sympy中定义的变量可以使用symbols(),如果定义单个变量可以使用symbol()函数，如x=symbol('x')。
symnols()函数可以接受一系列由空格分割的变量名字符串，并将其赋给相应的变量名。
例如x,y,z = symbols('x y z')。
limit()函数用来求极限，它有四个参数limit(f,x,x0,dir='+-)
f表示函数；
x表示要取极限的变量；
x0表示x趋近的数值；
dir表示取极限的方向，如dir=‘+’则表示取右极限，dir=‘-’取左极限，dir=‘+-’取双向左右极限。

【例2.8】设分段函数f(x),x<0,f(x)=x-1;x=0,f(x)=0;f(x)=x+1,证明x->0时，函数极限不存在。

```python
import sympy as sy

x=sy.symbols('x')
def f_left(x):return x-1
def f_right(x):return x+1
def f_mid(x):return 0
sy.limit(f_left(x),x,0,dir='-'),sy.limit(f_right(x),x,0,dir='+'),sy.limit(f_mid(x),x,0,dir='+-')
```

该分段函数在x->0时左右极限不等，故该函数在x->0时无极限。

【例2.19】说明数列2^0.5，(2+2^0.5)^.05,(2+(2+2^0.5)^0.5)^0.5……的极限存在。

解：注意到第n+1项=（2+第n项）^0.5，发现是递归机制的数列。

```python
import numpy as np
import matplotlib.pyplot as plt

def f(n):
    if n<=0:
        return 2**0.5
    # 一个函数调用自身，则这个函数就是递归函数。
    return (2.0+f(n-1))**0.5

x=[]
y=[]
for i in range(20):
    x.append(i)
    y.append(f(i))
print(np.array(x))
print(np.array(y))
plt.scatter(x, y,color='r',label='y=f(n)')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('Recursive Function f(n)')
plt.legend()        # 添加图例在最佳位置
plt.grid(True)      # 添加网格线
plt.show()
```

从输出的y值及散点图来看，随着n增大，数值越来越靠近2，从第15项开始后边的数值基本为2，故极限存在，其为2。

【例2.24】讨论当x->00时，(1-x^2n)/(1+x^2n)极限的连续性，若有断点，判断其类型。

```python
from sympy import symbols, Abs, oo

# 定义符号变量
x = symbols('x')
n = symbols('n', positive=True)  # n 是正整数

# 定义函数 f(x)
def f(x):
    # 手动分析 x 的取值范围
    if Abs(x) < 1:  # |x| < 1
        return 1
    elif Abs(x) > 1:  # |x| > 1
        return -1
    else:  # x == 1 或 x == -1
        return 0

# 定义 epsilon
epsilon = 1e-5

# 计算函数值
print("f(1 - epsilon):", f(1 - epsilon))  # 应该接近 1
print("f(1):", f(1))  # 应该为 0
print("f(1 + epsilon):", f(1 + epsilon))  # 应该接近 -1

print("f(-1 - epsilon):", f(-1 - epsilon))  # 应该接近 -1
print("f(-1):", f(-1))  # 应该为 0
print("f(-1 + epsilon):", f(-1 + epsilon))  # 应该接近 1
```

书上的源代码运行是错误的，通过观察当x=1和x=-1时，函数值为0，因此分别从这两个特殊值去求左右极限
只有左右极限存在且相等时，该点连续。
通过求解过程发现在(-oo，-1)和(1,oo)区间函数连续，函数在x=1和x=-1时，存在断点即函数不连续。

为了更清楚的看清楚函数的情况，画图如下：

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义极限函数 f(x)
def f(x):
    if np.abs(x) < 1:
        return 1
    elif np.abs(x) > 1:
        return -1
    else:  # x == 1 或 x == -1
        return 0

# 生成 x 值
x_values = np.linspace(-2, 2, 1000)  # 从 -2 到 2，生成 1000 个点
y_values = [f(x) for x in x_values]  # 计算对应的 y 值

# 绘制图形
plt.plot(x_values, y_values, label=r'$f(x) = \lim_{n \to \infty} \frac{1 - x^{2n}}{1 + x^{2n}}$', color='blue')

# 标记特殊点
plt.scatter([-1, 1], [f(-1), f(1)], color='red', zorder=5)  # 在 x = -1 和 x = 1 处标记点
plt.annotate('(-1, 0)', xy=(-1, 0), xytext=(-1.2, 0.2), fontsize=12, color='red')  # 标注 (-1, 0)
plt.annotate('(1, 0)', xy=(1, 0), xytext=(0.8, 0.2), fontsize=12, color='red')  # 标注 (1, 0)

# 添加标题和标签
plt.title(r'Graph of $f(x) = \lim_{n \to \infty} \frac{1 - x^{2n}}{1 + x^{2n}}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # 绘制 x 轴
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # 绘制 y 轴
plt.legend()
plt.grid(True)

# 显示图形
plt.show()
```

以下是绘制函数$f(x) = \frac{\log_a(1 + x)}{x}$在x->0附近的图像，并验证极限=1/ln(a)的代码。

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置支持中文的字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用 SimHei 字体（黑体）
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 定义函数 f(x)
def f(x, a):
    return np.log(1 + x) / (x * np.log(a))  # 使用自然对数 ln(1 + x) 和换底公式

# 定义 a 的值
a = 2  # 假设 a = 2，可以修改为其他正数

# 生成 x 值
x_values = np.linspace(-0.99, 1, 1000)  # x 从 -0.99 到 1，避免 x = -1 时 log(1 + x) 无定义
x_values = x_values[x_values != 0]  # 去掉 x = 0，避免除以 0
y_values = f(x_values, a)  # 计算对应的 y 值

# 计算极限值
limit_value = 1 / np.log(a)  # 理论极限值

# 绘制图形
plt.plot(x_values, y_values, label=r'$f(x) = \frac{\log_a(1 + x)}{x}$', color='blue')
plt.axhline(limit_value, color='red', linestyle='--', label='极限值 $\\frac{1}{\\ln a}$')  # 绘制极限值
plt.scatter(0, limit_value, color='red', zorder=5)  # 标记极限点

# 添加标题和标签
plt.title(r'函数图像 $f(x) = \frac{{\log_a(1 + x)}}{{x}}$ (a = {})'.format(a), fontsize=12)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # 绘制 x 轴
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # 绘制 y 轴
plt.legend(fontsize=12)
plt.grid(True)

# 显示图形
plt.show()
```

其图像方式展示

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
import math

# 定义符号变量和函数
x = sy.symbols('x')
a = 2  # 你可以将 a 替换为任何正数

# 定义表达式
expr = (a**x - 1) / x

# 计算极限
limit_expr = sy.limit(expr, x, 0)

# 打印极限结果
print(f"当 x -> 0 时，(a^x - 1)/x 的极限为: {limit_expr}")
print(f"ln(a) 的值为: {math.log(a)}")

# 绘制函数图像
def f(x_val, a_val):
    return (a_val**x_val - 1) / x_val

# 定义 x 的范围（避免 x = 0）
x_vals = np.linspace(-5, 5, 400)
x_vals = x_vals[x_vals != 0]  # 移除 x = 0 的点

# 计算 y 值
y_vals = f(x_vals, a)

# 绘制图像
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$\frac{a^x - 1}{x}$")
plt.axhline(y=math.log(a), color='r', linestyle='--', label=r"$\ln(a)$")  # 极限值
plt.axvline(x=0, color='k', linestyle=':', alpha=0.5)  # x = 0 的参考线
plt.title(f"Behavior of $\\frac{{a^x - 1}}{{x}}$ as $x \\to 0$ (a = {a})")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# 定义符号变量
x = sp.symbols('x')

# 定义表达式
expr = (1 + 2*x)**(3/sp.sin(x))

# 计算极限
limit_expr = sp.limit(expr, x, 0)

# 打印极限结果
print(f"当 x -> 0 时，(1 + 2x)^(3/sin(x)) 的极限为: {limit_expr}")

# 绘制函数图像
def f(x_val):
    return (1 + 2*x_val)**(3/np.sin(x_val))

# 定义 x 的范围（避免 x = 0）
x_vals = np.linspace(-0.1, 0.1, 400)
x_vals = x_vals[x_vals != 0]  # 移除 x = 0 的点

# 计算 y 值
y_vals = [f(xi) for xi in x_vals]

# 绘制图像
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$(1 + 2x)^{3/\sin(x)}$")
plt.axhline(y=np.exp(6), color='r', linestyle='--', label=f"Limit: $e^6$")  # 极限值
plt.axvline(x=0, color='k', linestyle=':', alpha=0.5)  # x = 0 的参考线
plt.title(r"Behavior of $(1 + 2x)^{3/\sin(x)}$ as $x \to 0$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# 定义符号变量
x = sp.symbols('x')

# 定义表达式
expr = (sp.exp(1/x) - 1) / (sp.exp(1/x) + 1)

# 计算极限
limit_right = sp.limit(expr, x, 0, dir='+')  # x -> 0^+
limit_left = sp.limit(expr, x, 0, dir='-')  # x -> 0^-

# 打印极限结果
print(f"当 x -> 0^+ 时，极限为: {limit_right}")
print(f"当 x -> 0^- 时，极限为: {limit_left}")

# 绘制函数图像
def f(x_val):
    return (np.exp(1/x_val) - 1) / (np.exp(1/x_val) + 1)

# 定义 x 的范围（正方向和负方向）
x_vals_positive = np.linspace(0.01, 1, 400)  # x -> 0^+
x_vals_negative = np.linspace(-1, -0.01, 400)  # x -> 0^-

# 计算 y 值
y_vals_positive = [f(xi) for xi in x_vals_positive]
y_vals_negative = [f(xi) for xi in x_vals_negative]

# 绘制图像
plt.figure(figsize=(10, 6))
plt.plot(x_vals_positive, y_vals_positive, label=r"$x \to 0^+$", color='blue')
plt.plot(x_vals_negative, y_vals_negative, label=r"$x \to 0^-$", color='red')
plt.axhline(y=1, color='green', linestyle='--', label="Limit: 1 (x -> 0^+)")
plt.axhline(y=-1, color='purple', linestyle='--', label="Limit: -1 (x -> 0^-)")
plt.axvline(x=0, color='k', linestyle=':', alpha=0.5)  # x = 0 的参考线
plt.title(r"Behavior of $\frac{e^{1/x} - 1}{e^{1/x} + 1}$ as $x \to 0$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

【例2.30】当x->0时，求（（a^x+b^x+c^x)/3)^(1/x)的极限。（a>0,b>0,c>0）

```python
import sympy as sp

x = sy.symbols('x')
a,b,c = sy.symbols('a,b,c',positive=True)
y = ((a**x+b**x+c**x)/3)**(1/x)
lim_l = sy.limit(y,x,0,dir='-')
lim_r = sy.limit(y,x,0,dir='+')
if lim_l == lim_r:
    print("函数的极限为：",lim_l)
else:
    print("左右极限不相等！函数在此点没有极限。")
```

图形展示如下：

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# 定义符号变量
x = sp.symbols('x')
a, b, c = sp.symbols('a b c', positive=True)  # a, b, c > 0

# 定义表达式
expr = ((a**x + b**x + c**x) / 3)**(1/x)

# 计算极限
limit_expr = sp.limit(expr, x, 0)

# 打印极限结果
print(f"当 x -> 0 时，((a^x + b^x + c^x)/3)^(1/x) 的极限为: {limit_expr}")

# 绘制函数图像（固定 a, b, c 的值）
a_val = 2
b_val = 3
c_val = 5

# 定义数值函数
def f(x_val):
    return ((a_val**x_val + b_val**x_val + c_val**x_val) / 3)**(1/x_val)

# 定义 x 的范围（避免 x = 0）
x_vals = np.linspace(-0.1, 0.1, 400)
x_vals = x_vals[x_vals != 0]  # 移除 x = 0 的点

# 计算 y 值
y_vals = [f(xi) for xi in x_vals]

# 计算极限值
limit_val = (a_val * b_val * c_val)**(1/3)

# 绘制图像
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$\left( \frac{a^x + b^x + c^x}{3} \right)^{1/x}$")
plt.axhline(y=limit_val, color='r', linestyle='--', label=f"Limit: {limit_val:.4f}")  # 极限值
plt.axvline(x=0, color='k', linestyle=':', alpha=0.5)  # x = 0 的参考线
plt.title(f"Behavior of $\\left( \\frac{{a^x + b^x + c^x}}{{3}} \\right)^{{1/x}}$ as $x \\to 0$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 设置中文字体
rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 定义函数
def f(x):
    return (2*x - 1) * np.exp(1/x)

# 定义 x 的范围
x_vals_positive = np.linspace(0.01, 5, 400)  # x > 0
x_vals_negative = np.linspace(-5, -0.01, 400)  # x < 0

# 计算 y 值
y_vals_positive = [f(xi) for xi in x_vals_positive]
y_vals_negative = [f(xi) for xi in x_vals_negative]

# 定义斜渐近线
def asymptote(x):
    return 2*x + 1

# 计算渐近线的 y 值
y_asymptote_positive = [asymptote(xi) for xi in x_vals_positive]
y_asymptote_negative = [asymptote(xi) for xi in x_vals_negative]

# 绘制图像
plt.figure(figsize=(10, 6))
plt.plot(x_vals_positive, y_vals_positive, label=r"$y = (2x - 1)e^{1/x}$ ($x > 0$)", color='blue')
plt.plot(x_vals_negative, y_vals_negative, label=r"$y = (2x - 1)e^{1/x}$ ($x < 0$)", color='green')
plt.plot(x_vals_positive, y_asymptote_positive, label=r"$y = 2x + 1$ (斜渐近线)", color='red', linestyle='--')
plt.plot(x_vals_negative, y_asymptote_negative, color='red', linestyle='--')
plt.axvline(x=0, color='k', linestyle=':', label=r"$x = 0$ (垂直渐近线)")  # 垂直渐近线

# 设置坐标轴范围
plt.xlim(-5, 5)  # 调整 x 轴范围
plt.ylim(-10, 20)  # 调整 y 轴范围

# 添加标题和标签
plt.title("曲线 $y = (2x - 1)e^{1/x}$ 及其渐近线")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
